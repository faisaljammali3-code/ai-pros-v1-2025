### Dataset Class

```python
class HoursDataset(Dataset):
    
    def __init__(self,root_dir, transform=None):
        self.root_dir=root_dir
        self.transform=transform
        self.breed_map = {
            0: "Akhal-Teke",           
            1: "Appaloosa",            
            2: "Orlov Trotter",        
            3: "Vladimir Heavy Draft", 
            4: "Percheron",            
            5: "Arabian",              
            6: "Friesian"              
        }
        self.samples = self._load_samples()

    def __len__(self):
        return len(self.samples)


    def __getitem__(self, idx):
            image = self.retrieve_image(idx)

            if self.transform is not None:
                image = self.transform(image)

            label = self.samples[idx][1]

            return image, label

        
    def retrieve_image(self, idx):

            img_path = self.samples[idx][0]
            
            with Image.open(img_path) as img:
                image = img.convert("RGB")
            return image     
    def get_label_description(self, label):
            """
            Retrieves the text description for a given label index.
            """
            return self.breed_map.get(label, "Unknown")
    

    def _load_samples(self):

        samples = []
        for root, _, files in os.walk(self.root_dir):
            for file in files:
                if file.lower().endswith(('.png')):
                    prefix = file[:2]
                    
                    if prefix.isdigit():
                        label_idx = int(prefix) - 1
                        
                        if label_idx in self.breed_map:
                            full_path = os.path.join(root, file)
                            samples.append((full_path, label_idx))
        return samples

```
